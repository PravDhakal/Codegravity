```python
import pandas as pd
```

```python
df1 = pd.read_excel(r'C:\Users\Pravasis Dhakal\Desktop\project code gravity\CG Accommodation expense.xlsx')
```

```python
df1
```

```python
df1.dropna(how='all', axis=0, inplace=True)
```

```python
df1.dropna(how='all', axis=1, inplace=True)
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
df1.head()
```

```python
df1.columns = ["Apartment", "Tenant Type"] + list(df1.columns[2:])
```

```python
df1
```

```python
df1["Apartment"].fillna(method="ffill", inplace=True)
```

```python
df1
```

```python
for col in df1.columns[2:]:
    df1[col] = pd.to_numeric(df1[col], errors="coerce")
```

```python
df1
```

```python

df_long = df1.melt(id_vars=["Apartment", "Tenant Type"], var_name="Date_Expense", value_name="Amount")
```

```python
df_long
```

```python
df_long["Date"] = pd.to_datetime(df_long["Date_Expense"].str.extract(r'(\d{4}-\d{2}-\d{2})')[0])
```

```python

df_long
```

```python
df_long["Expense Type"] = df_long["Date_Expense"].str.extract(r'([A-Za-z]+)$')[0]
```

```python

df_long
```

```python
df_long.drop(columns=["Date_Expense"], inplace=True)
```

```python

df_long
```

```python
df_long.dropna(subset=["Date", "Expense Type"], inplace=True)
```

```python

df_long
```

```python
df_long["Amount"] = pd.to_numeric(df_long["Amount"], errors="coerce")
```

```python

df_long
```

```python
df_long.to_excel("Cleaned_CG_Accommodation_Expense.xlsx", index=False)
```

```python
print("Data cleaning and transformation completed. Saved as 'Cleaned_CG_Accommodation_Expense.xlsx'.")
```
